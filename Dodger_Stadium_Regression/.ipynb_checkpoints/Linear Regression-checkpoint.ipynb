# Linear Regression for Dodger Stadium Attendance

Below is the script I wrote to produce the Dodger attendance regression.

### Data Sources
- The attendance figures and game information come from baseball-reference.com

- The weather data comes from OpenWeatherMap

- The bobble head data was taken from the Dodgers' website

## Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in data set that includes baseball-reference and OpenWeatherMap data

df = pd.read_csv("Data/baseball_data.csv")
```

```python
# Convert date column to string

df["Date"] = df["Date"].astype('str')

# Create variables for day of the week

for day in ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']:
    list_ = df["DOW"] == day
    list_encoded = []

    for element in list_:
        if element:
            list_encoded.append(1)
        else:
            list_encoded.append(0)
    df["encoded_" + day] = list_encoded
    
# Create a column in df called seasons that represents each season with a number

seasons = []

for date in df["Date"]:
    if date[0:4] == '2017':
        seasons.append(7)
    elif date[0:4] == '2016':
        seasons.append(6)
    elif date[0:4] == '2015':
        seasons.append(5)
    elif date[0:4] == '2014':
        seasons.append(4)
    elif date[0:4] == '2013':
        seasons.append(3)
    elif date[0:4] == '2012':
        seasons.append(2)
    elif date[0:4] == '2011':
        seasons.append(1)
    else:
        print("error")
        
df["Seasons"] = seasons

# Convert Attendance values to floats

df["Attendance"] = df["Attendance"].str.replace(',', '').astype(float)

# Convert Date column to datetime

df["Date"] = pd.to_datetime(df["Date"])


```

```python
# Sort data by date of game

df = df.sort_values(by='Date')
```

## Linear Regression Analysis

```python
from sklearn import linear_model, preprocessing
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from yellowbrick.regressor import ResidualsPlot
import seaborn as sns
```

```python
# Pick features to include in first regression analysis and assign independent vars to x and 
# dependent var to Y

features = ['Is_Weekend','Month','Win_Per',
       'Is_Day', 'Bobble_head', 'Precip', 'Temp',
       'Seasons', 'Game_ID']

X = df.loc[:, features]

Y = df["Attendance"]
```

```python
# Normalize data with MinMax technique

scaler = preprocessing.MinMaxScaler()

scaled_df = scaler.fit_transform(X)

scaled_df = pd.DataFrame(scaled_df, columns=features)
```

```python
# Include the below variables in regression model

features = ['Is_Weekend','Month','Win_Per',
       'Is_Day', 'Bobble_head', 'Precip', 'Temp',
       'Seasons', ]

X = scaled_df.loc[:, features] 
```

```python
# Fit regression and output regression statistics

import statsmodels.api as sm
from scipy import stats

X2 = sm.add_constant(X)
est = sm.OLS(list(Y), X2)
est2 = est.fit()
print(est2.summary())

```

```python
# Fit model to data using Sklearn library

lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
predictions = lm.predict(X)
```

```python
# Plot model predictions overs actual attendance

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:" + str(fig_size))
 
# Set figure width to 12 and height to 9
fig_size[0] = 14
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.title("Linear Regression Performance")
plt.plot(df["Date"], Y, '-b', label = 'Actuals')
plt.plot(df["Date"], predictions, color = "orange", label = 'Predictions')

plt.ylabel("Attendance")
plt.xlabel("Date of Game")
plt.legend(loc='upper right')
plt.show()
```

```python
# Zoom into one season of Predictions vs Actuals graph

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:" + str(fig_size))
 
# Set figure width to 12 and height to 9
fig_size[0] = 14
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.title("Linear Regression Performance")
plt.ylabel("Attendance")
plt.xlabel("Date")
plt.plot(df["Date"][90:160], Y[90:160], '-b', label = 'Actuals')
plt.plot(df["Date"][90:160], predictions[90:160], color = "orange", label = 'Predictions')
plt.legend(loc='upper right')
plt.show()
```

```python
# Residual plot for first linear regression

resid = LinearRegression()
visualizer = ResidualsPlot(resid)

visualizer.fit(X,Y)  
visualizer.score(X,Y)  
visualizer.poof()   
```

#### Observations:
- Month has a high p-value of .663 and may not belong in this model.
- There is a linear trend in the residuals chart that indicates the model may be missing a variable. 

## Regression Including Away Team Info

This second version of the linear regression includes a dummy variable for each of the Dodgers' opponents. Perhaps certain teams reliably attract more fans than other teams.

```python
# Create dummy variables for each opponent of Dodgers

for number in np.arange(1,30):
    oppo = df["Opponent_Number"] == number
    oppo_encoded = []

    for element in oppo:
        if element:
            oppo_encoded.append(1)
        else:
            oppo_encoded.append(0)
    df["encoded_team_" + str(number)] = oppo_encoded
```

```python
# Select new features

features = ['Is_Weekend','Month','Win_Per',
       'Is_Day', 'Bobble_head', 'Precip', 'Temp',
       'Seasons','Game_ID',  'encoded_team_1', 'encoded_team_2',
       'encoded_team_3', 'encoded_team_4', 'encoded_team_5', 'encoded_team_6',
       'encoded_team_7', 'encoded_team_8', 'encoded_team_9', 'encoded_team_10',
       'encoded_team_11', 'encoded_team_12', 'encoded_team_13',
       'encoded_team_14', 'encoded_team_15', 'encoded_team_16',
       'encoded_team_17', 'encoded_team_18', 'encoded_team_19',
       'encoded_team_20', 'encoded_team_21', 'encoded_team_22',
       'encoded_team_23', 'encoded_team_24', 'encoded_team_25',
       'encoded_team_26', 'encoded_team_27', 'encoded_team_28',
       'encoded_team_29']

X = df.loc[:, features]
Y = df["Attendance"]
```

```python
# Normalize data with MinMax technique

scaler = preprocessing.MinMaxScaler()

scaled_df = scaler.fit_transform(X)

scaled_df = pd.DataFrame(scaled_df, columns=features)

X = scaled_df

X = X.drop(['Game_ID'], axis = 1)
```

```python
# Fit regression and output regression statistics

X2 = sm.add_constant(X)
est = sm.OLS(list(Y), X2)
est2 = est.fit()
print(est2.summary())
```

```python
# Train regression model with away team information

lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
predictions = lm.predict(X)
```

```python
# Plot performance of regression

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:" + str(fig_size))
 
# Set figure width to 12 and height to 9
fig_size[0] = 14
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size


plt.title("Linear Regression Performance")
plt.plot(df["Date"], Y, '-b', label = 'Actuals')
plt.plot(df["Date"], predictions, color = "orange", label = 'Predictions')

plt.ylabel("Dodger Stadium Attendance")
plt.xlabel("Date")
plt.legend(loc='upper right')
plt.savefig('overall_performance.png')
plt.show()


```

```python
# Zoom in on one year of residual predictions

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:" + str(fig_size))
 
# Set figure width to 12 and height to 9
fig_size[0] = 14
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.plot(df["Date"][90:160], Y[90:160], '-b', label = 'Actuals')
plt.plot(df["Date"][90:160], predictions[90:160], color = "orange", label = 'Predictions')

plt.title("Linear Regression Performance")
plt.ylabel("Dodger Stadium Attendance")
plt.xlabel("Date")
plt.legend(loc='upper right')
plt.savefig('year_performance.png')

plt.show()
```

```python
# Generate Residual chart and coefficient valuea

ridge = LinearRegression()
visualizer = ResidualsPlot(ridge)



visualizer.fit(X,Y)  
visualizer.score(X,Y)  
visualizer.poof()    

coef = list(zip(lm.coef_,features))

for element in coef:
    print(element[1] + " : " + str(element[0])) 
```

```python
# Drop all coefficients that are not statistically significant, retrain model

X = X.drop(['Month',
        'encoded_team_1', 'encoded_team_2',
       'encoded_team_3', 'encoded_team_4',  'encoded_team_6',
       'encoded_team_7', 'encoded_team_8', 'encoded_team_9', 'encoded_team_10',
       'encoded_team_11', 'encoded_team_12', 
       'encoded_team_14', 'encoded_team_15', 'encoded_team_16',
       'encoded_team_17', 'encoded_team_18', 'encoded_team_19',
       'encoded_team_20',  'encoded_team_22',
       'encoded_team_23', 'encoded_team_24', 'encoded_team_25',
       'encoded_team_26',  'encoded_team_28'], axis = 1)

# Replace encoded_team column name with actual team name

X.columns = ['Is_Weekend', 'Win_Per', 'Is_Day', 'Bobble_head', 'Precip', 'Temp',
       'Seasons', 'SF_Giants', 'LA_Angels', 'BOS_Red_Soxs',
       'NY_Yankees', 'FL_Marlins']

# Output the residual statistics

X2 = sm.add_constant(X)
est = sm.OLS(list(Y), X2)
est2 = est.fit()
print(est2.summary())
```

```python
# Generate Residual chart and coefficient valuea

ridge = LinearRegression()
visualizer = ResidualsPlot(ridge)


visualizer.fit(X,Y)  
visualizer.score(X,Y)  
visualizer.poof() 
```

#### Observations:
-  The adjusted R squared of .37 is an improvment from the .32 of the first regression.
-  P-values of the coefficients are low.
-  All the signs of the coefficients make sense except for Temp. The model indicates that attendance increases as the temperature rises. Perhaps Temp is a proxy for game number and game number has a relationship with attendance.

## Regression with Game_ID instead of Temp

I want to see if adding Game_ID as a variable improves the model. Month did not have a clear relationship with attendance but perhaps Game_ID will.


```python
# Adding Temp and Game_ID to the same model may produce issues 
# The graph below shows their collinearity

X["Game_ID"] = scaled_df.loc[:, "Game_ID"]

plt.scatter(X["Game_ID"], X["Temp"])
plt.title("Game_ID vs. Temp")
plt.xlabel("Game_ID")
plt.ylabel("Temp")
plt.show()
```

```python
# Print out one model with both Game_ID and Temp and one model with only Game_ID

print( "Statistics with both Temp and Game_ID\n")


X2 = sm.add_constant(X)
est = sm.OLS(list(Y), X2)
est2 = est.fit()
print(est2.summary())

X = X.drop(["Temp"], axis = 1)

print( "\nStatistics with only Game_ID\n")
X2 = sm.add_constant(X)
est = sm.OLS(list(Y), X2)
est2 = est.fit()
print(est2.summary())

```

#### Observations
- The p-value of Game_ID is quite high when Temp and Game_ID are both included in the model but it drops to .021 when Temp is removed. This suggests these variables are collinear.
- Replacing Temp with Game_ID hurts the adjusted R squared score by .01 but produces coefficients that make more sense. Game_ID should be used instead of Temp. 

# Adding New Features to Regression

Here I am adding two new independent variables to the model (an opening day variable and a Kershaw variable). As can be seen in the regression performance graphs above, the model consistently underestimates the attendance on opening day so a opening day variable should improve the accuracy of the model. I am adding a dummy variable called Kershaw that reflects whether or not Clayton Kershaw is the starting pitcher for the game. He is the Dodgers' star player so attendance might increase when he pitches.

```python
# import new independent variable data

new_data = pd.read_csv("Data/new_data_2.csv")
```

```python
# Add new vars to df

X["Kershaw"], X["Opening_Day"] = new_data["Kershaw "], new_data["Opening_Day"]
```

```python
# View new model statistics

X2 = sm.add_constant(X)
est = sm.OLS(list(Y), X2)
est2 = est.fit()
print(est2.summary())
```

```python
# Drop Kershaw variable because of its high P-value

X = X.drop(["Kershaw"], axis = 1)

```

```python
# Fit model to data 
lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
predictions = lm.predict(X)
```

```python
# Generate Residual chart

ridge = LinearRegression()
visualizer = ResidualsPlot(ridge)


visualizer.fit(X,Y)  
visualizer.score(X,Y)  
visualizer.poof()    

features.append("Opening_Day")

```

#### Observations
